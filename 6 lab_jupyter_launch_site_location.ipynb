```python
# Task 2: Mark the success/failed launches for each site on the map
# Import MarkerCluster if not already done
from folium.plugins import MarkerCluster

# Create a map centered on the average coordinates of all launches
center_lat = spacex_df['Lat'].mean()
center_lon = spacex_df['Long'].mean()
map_clusters = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Initialize a MarkerCluster object
marker_cluster = MarkerCluster().add_to(map_clusters)

# Iterate through each launch and add a colored marker
for index, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    site = row['Launch Site']
    outcome_class = row['class']
    
    # Set marker color based on success/failure
    color = 'green' if outcome_class == 1 else 'red'
    outcome_text = 'Success' if outcome_class == 1 else 'Failure'
    
    # Popup with launch site and outcome
    popup_text = f"<b>{site}</b><br>Outcome: {outcome_text}"
    
    # Add marker to the cluster
    folium.Marker(
        location=coord,
        popup=popup_text,
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(marker_cluster)

# Display the map
map_clusters
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math
import folium
from folium.plugins import MarkerCluster

# ----- 1. Define Haversine function -----
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points
    on the Earth (specified in decimal degrees).
    Returns distance in kilometers.
    """
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    r = 6371  # Radius of Earth in kilometers
    return c * r

# ----- 2. Ensure data is loaded -----
# If `spacex_df` is missing, reload it
try:
    spacex_df.head()
except NameError:
    from js import fetch
    import io
    URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
    resp = await fetch(URL)
    spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
    spacex_df = pd.read_csv(spacex_csv_file)

# Get unique launch sites
launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates().reset_index(drop=True)

# ----- 3. Create or reuse a map -----
# If you already have a `site_map` from Task 2, use it; otherwise create a new one.
try:
    site_map  # check if exists
except NameError:
    site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)
    # Also add marker cluster for launches (optional, but you may want it)
    marker_cluster = MarkerCluster().add_to(site_map)
    for index, record in spacex_df.iterrows():
        marker = folium.Marker(
            location=[record['Lat'], record['Long']],
            popup=f"<b>{record['Launch Site']}</b><br>Outcome: {'Success' if record['class']==1 else 'Failure'}",
            icon=folium.Icon(color='white', icon_color='green' if record['class']==1 else 'red', icon='info-sign')
        )
        marker_cluster.add_child(marker)

# ----- 4. Proximity points (coastline, cities) -----
proximities = {
    'Coastline_CCAFS': (28.4, -80.5),
    'Coastline_KSC': (28.6, -80.6),
    'Coastline_VAFB': (34.6, -120.5),
    'Orlando_City': (28.5383, -81.3792),
    'LA_City': (34.0522, -118.2437)
}

# ----- 5. Add distance markers and lines -----
for idx, site_row in launch_sites_df.iterrows():
    site_name = site_row['Launch Site']
    site_lat = site_row['Lat']
    site_lon = site_row['Long']
    
    for prox_name, (p_lat, p_lon) in proximities.items():
        dist = haversine(site_lat, site_lon, p_lat, p_lon)
        
        # Add a marker at the proximity point
        folium.Marker(
            location=[p_lat, p_lon],
            popup=f"{prox_name}<br>Distance from {site_name}: {dist:.2f} km",
            icon=folium.Icon(color='purple', icon='flag')
        ).add_to(site_map)
        
        # Draw a line connecting site and proximity point
        folium.PolyLine(
            locations=[[site_lat, site_lon], [p_lat, p_lon]],
            color='gray',
            weight=1,
            opacity=0.5
        ).add_to(site_map)

# ----- 6. Display the map -----
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
